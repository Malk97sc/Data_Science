### DEFINIENDO LA FRECUENCIA
La frecuencia debe ser constante para poder hacer analisis de series temporales.

```python
df_comp=df_comp.asfreq('D')

# h indicara que se registra cada hora
# w datos semanales
# d datos diarios
# m datos mensuales
# a datos anuales (annual)

#PROBAR B: business days
```

```python
df_comp.head(50)

# crea los dias que faltan para tener siempre el mismo periodo de tiempo
# NaN not a number
# No se tienen datos de los fines de semana que no funcionan las bolsas - o por vacaciones - o por errores ...
# no tiene sentido considerarlo ya que no hay datos de fin de semana o para dias no habiles
# no se tienen datos
```

```python
df_comp=df_comp.asfreq('B')

# la letra b se refiere a business days. Deja fuera dias de trabajo y genera la serie temporal solo con los dias que deben de tener valor
# todavia pueden faltar si hubo errores!!
# la frecuencia sigue siendo la misma por dia -business day - de viernes a lunes no hay ningun problema
```

```python
df_comp.head(100)
```

```python
df_comp=df_comp.asfreq('h')
```

### VALORES FALTANTES (VER  DETALLES EN PPT)

```python
df_comp=df_comp.asfreq('b')

df_comp.isna().sum()

# despues de ponr la frecuencia adecuada, vemos que hay 8 que faltan
# faltan 8 fechas para las 4 variables
```

```python
df_comp.isna()
```

```python
df_comp.isna().sum()
```

```python
print("\nRows containing missing values:")
print(df_comp[df_comp.isnull().any(axis=1)])
```

```python
df_comp
```

```python
df_comp.spx=df_comp.spx.fillna(method='ffill')

# el metodo fillna rellena los NaN - los valores faltantes.
#el metodo ffill : front filling - asigna al valor faltante el valor del periodo posterior (el de enfrente)
```

```python
df_comp.isna().sum()
```

```python
df_comp.ftse=df_comp.ftse.fillna(method='bfill')

# back filling - lo rellena con el periodo anterior (el dia anterior en este caso)
```

```python
df_comp.dax=df_comp.dax.fillna(value = df_comp.dax.mean())

#otra forma es asignar el valor medio a los valores faltantes
#en general para series temporales esto no tiene sentido - solo seria apropiado si fluctua alrededor de la media 
# si hay tendencia , al no ser estable, no tiene sentido ...
```

```python
df_comp.nikkei=df_comp.dax.fillna(method='bfill')
```

```python
df_comp.isna().sum()

#ya no falta ninguno
```
